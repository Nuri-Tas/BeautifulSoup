```python
import requests
from bs4 import BeautifulSoup

from selenium import webdriver 
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
from time import sleep
```

```python
driver_path = Service(r"C:\Users\Lenovo\Desktop\chromedriver.exe")
driver = webdriver.Chrome(service=driver_path)
driver.get('https://store.unionlosangeles.com/collections/outerwear')

brands = []
df = pd.DataFrame(columns=['brand', 'title', 'price', 'image'])


while True:
    for i in range(1, 31):
        try:
            sleep(3)
            title_path = '//*[@id="isp_search_results_container"]/li[' + str(i) + ']/div[2]/div[4]'
            title = driver.find_element(By.XPATH, title_path).text

            brand_path = '//*[@id="isp_search_results_container"]/li[' + str(i) + ']/div[2]/div[3]'
            brand = driver.find_element(By.XPATH, brand_path).text

            price_path = '//*[@id="isp_search_results_container"]/li[' + str(i) + ']/div[2]/div[5]/span'
            price = driver.find_element(By.XPATH, price_path).text

            sleep(1)
            img_path = '//*[@id="isp_search_results_container"]/li[' + str(i) + ']/div[1]/a/img'
            img = driver.find_element(By.CLASS_NAME, 'isp_product_image').get_attribute('src')

            #df = df.append({'brand':brand, 'title':title, 'price':price, 'image':img}, ignore_index=True)
            new_index = len(df)
            df.loc[new_index] = [brand, title, price, img]
        except NoSuchElementException:
            pass
    
    sleep(3)
    try:
        next_path = '//*[@id="isp_pagination_anchor"]/ul/li[8]/a'    
        next_page = driver.find_element(By.XPATH, next_path)
        next_page.send_keys(Keys.ENTER)
    except ElementNotInteractableException:
        print('DONE')
        break
        
```

Convert the price column into float

```python
def dollar(entry):
    entry = entry.replace('$', '')
    return float(entry)
```

```python
df.price = df.price.apply(lambda x: dollar(x))
df.head()
```
