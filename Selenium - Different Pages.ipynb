```python
from selenium.common.exceptions import NoSuchElementException

driver = webdriver.Chrome(service=chrome_driver_path)
driver.get('https://www.dgap.de/')

news = []
performance = []
links = []
desc = []
for i in range(1, 21):
    path = '//*[@id="content"]/div[1]/div[1]/div[1]/table/tbody/tr[' + str(i) + ']/td[2]/a/strong'
    new = driver.find_element(By.XPATH, path).text
    news.append(new)
    
    link_path = '//*[@id="content"]/div[1]/div[1]/div[1]/table/tbody/tr[' + str(i) + ']/td[2]/a'
    news_link = driver.find_element(By.XPATH, link_path).get_attribute('href')
    links.append(news_link)
    
    
    desc_path = '//*[@id="content"]/div[1]/div[1]/div[1]/table/tbody/tr[' + str(i) + ']/td[2]/a' 
    desc_link = driver.find_element(By.XPATH, desc_path).text
    desc.append(desc_link)
    
    try:
        perf_path = '//*[@id="content"]/div[1]/div[1]/div[1]/table/tbody/tr[' + str(i) + ']/td[3]/div[1]/span'
        perf = driver.find_element(By.XPATH, perf_path).text
        performance.append(perf)
    except NoSuchElementException:
        performance.append('No information')
        

```

```python
date = [''.join(date) for item in news  for date in item.split(' ')[0:2]]
date = [' '.join(date[i:i + 2]) for i in range(0, len(date), 2)]
date
```

```python
news
```

```python
splitted_news = [item.split(' ')[2:] for item in news]
titles = []
for item in splitted_news:
    title = ''
    for element in item:
        title += element
    titles.append(title)    
        
```

```python
performance
```

```python
import pandas as pd 

df = pd.DataFrame({'news': titles, 'performance':performance, 'description':desc, 'news_link':links, 'date':date})
```

Get the inmate information from the website https://redrock.clarkcountynv.gov/ccdcincustody/inCustodySearch.aspx Only, look at the inmates whose name start with 'GA' and filter out 

```python
from time import sleep
```

```python
driver = webdriver.Chrome(service=chrome_driver_path)
driver.get('https://redrock.clarkcountynv.gov/ccdcincustody/inCustodySearch.aspx')

inmates_name = 'GA'
box = driver.find_element(By.XPATH, '//*[@id="txtName"]')
box.send_keys(inmates_name)

submit_box = driver.find_element(By.XPATH, '//*[@id="SearchName"]')
submit_box.click()

names = []
dates = []
for i in range(2, 73):
    sleep(0.5)
    id_path  = '//*[@id="DataGrid1"]/tbody/tr[' + str(i) + ']/td[1]/input'
    id_box = driver.find_element(By.XPATH, id_path)
    id_box.click()
    

    try:
        sleep(1)
        arrest_name = driver.find_element(By.XPATH, '/html/body/table[2]/tbody/tr[1]/td[2]').text
        print(arrest_name)
        names.append(arrest_name)
    except NoSuchElementException:
        names.append('no name')
        
    try:   
        sleep(1)
        arrest_date = driver.find_element(By.XPATH, '/html/body/table[2]/tbody/tr[3]/td[2]').text
        print(arrest_date)
        dates.append(arrest_date)
    except NoSuchElementException:
        dates.append('no date')
    
    driver.back()
    
print('DONE')    
```

```python
websites = []
for i in range(100):

    url = df.source_link[i]
    text = requests.get(url).text
    soup = BeautifulSoup(text, 'lxml')
    class_ = []
    for item in soup.find_all('a',{'class':'css-1um3nx', 'role':'link'}):
        class_.append(item.string)
    exists = False    
    for item in class_:
        
        try:
            if '.com' in item:
                print(item)
                websites.append(item)
                exists = True
        except TypeError:
            pass
    if not exists:
        websites.append('not found')
```
